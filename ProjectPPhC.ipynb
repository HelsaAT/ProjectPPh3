## Data wrangling

### Subtask:
Prepare the data for analysis and visualization.


**Reasoning**:
Convert the 'tahun' column to datetime objects, encode categorical features, create new features (profit margin), and scale numerical features using MinMaxScaler.



```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime

# Convert 'tahun' to datetime objects
try:
    df['tahun'] = pd.to_datetime(df['tahun'], format='%Y')
except ValueError:
    print("Error converting 'tahun' to datetime. Check the format.")

# One-hot encode categorical features
categorical_cols = ['skenario', 'tax_holiday_status', 'kategori']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Feature Engineering: Profit Margin
df['profit_margin'] = (df['laba_kotor'] / df['pendapatan']) * 100

# Feature Scaling: Using MinMaxScaler for numerical features
numerical_cols_to_scale = ['pendapatan', 'beban_operasional', 'penyusutan', 'laba_kotor', 'tax_rate',
                          'pajak_terutang', 'jumlah_aset', 'total_nilai_perolehan',
                          'rata_rata_umur_ekonomis', 'total_penyusutan_tahunan']
scaler = MinMaxScaler()
df[numerical_cols_to_scale] = scaler.fit_transform(df[numerical_cols_to_scale])

display(df.head())
```

## Data analysis

### Subtask:
Analyze the prepared data to gain insights.


**Reasoning**:
Calculate descriptive statistics, analyze correlations, investigate profit margin distribution across categories, and analyze the trend of profit margin over time.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
print("Descriptive Statistics:")
display(df.describe())

# Focus on 'profit_margin'
print("\nDescriptive Statistics for 'profit_margin':")
display(df['profit_margin'].describe())

# 2. Correlation Analysis and Heatmap
print("\nCorrelation Analysis:")
correlation_matrix = df.corr()
display(correlation_matrix)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()

# 3. Profit Margin Distribution across Categories
categorical_cols = ['skenario_Optimis', 'skenario_Pessimis', 'tax_holiday_status_Ya', 'kategori_Furniture', 'kategori_Kendaraan', 'kategori_Mesin', 'kategori_Peralatan']

for col in categorical_cols:
    print(f"\nProfit Margin Distribution by {col}:")
    display(df.groupby(col)['profit_margin'].agg(['mean', 'std']))

    plt.figure(figsize=(8, 6))
    sns.boxplot(x=col, y='profit_margin', data=df, palette='Set3')
    plt.title(f'Profit Margin Distribution by {col}')
    plt.show()

# 4. Trend of Profit Margin over Time
print("\nProfit Margin Trend over Time:")
average_profit_margin_by_year = df.groupby('tahun')['profit_margin'].mean()
display(average_profit_margin_by_year)

plt.figure(figsize=(10, 6))
plt.plot(average_profit_margin_by_year.index, average_profit_margin_by_year.values, marker='o', linestyle='-', color='teal')
plt.xlabel('Year')
plt.ylabel('Average Profit Margin')
plt.title('Trend of Average Profit Margin Over Time')
plt.grid(True)
plt.show()
```
